```python
from PIL import Image

img_path = '../tom.jpg'

# read
im = Image.open(img_path)
print(im.format, im.size, im.mode)
```

```python
# 默认调用本机的图片查看器打开 bmp 图片
im.show()
im
```

```python
size = (80, 80)
im.thumbnail(size)  # 创建缩略图
im
```

```python
im.size  # 保持了长边的一致
```

```python
im.save(img_path.replace('.jpg','_s.jpg')) # 保存缩略图为原图 jpg 格式
```
